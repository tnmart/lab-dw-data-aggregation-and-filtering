```python
#Starting by importing the libraries, functions and loading the CSV file:

import pandas as pd
import functions

df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv")

display(df)
```

```python
# Cleaning the column names and dropping the 'unnamed:_0' column which was a copy of the label column
df=functions.cleaning_column_names(df)
df=df.drop('unnamed:_0', axis=1)

df.head()
```

```python
#Checking for duplicates and cleaning them:

display(df.duplicated().sum())
df=functions.dealing_with_duplicates(df)
display(df.duplicated().sum())
```

```python
#Checking for entire rows of missing values:
df.isna().all(axis=1).sum()
```

```python
#Checking for percentage of missing values in each column:

display(df.isna().sum()/len(df)*100)
```

```python
#Checking the column data types.
display(df.dtypes)

df['effective_to_date']=pd.to_datetime(df['effective_to_date'])

display(df.dtypes)
```

```python
# Creating new functions to clean the missing values:

def fill_nan_median(dataframe: pd.DataFrame, col_name: str)-> pd.DataFrame:
    dataframe2=dataframe.copy()
    dataframe2[col_name]=dataframe2[col_name].fillna(dataframe2[col_name].median())
    return dataframe2

def fill_nan_mean(dataframe: pd.DataFrame, col_name: str)-> pd.DataFrame:
    dataframe2=dataframe.copy()
    dataframe2[col_name]=dataframe2[col_name].fillna(dataframe2[col_name].mean())
    return dataframe2

def fill_nan_mode(dataframe: pd.DataFrame, col_name: str)-> pd.DataFrame:
    dataframe2=dataframe.copy()
    dataframe2[col_name]=dataframe2[col_name].fillna(dataframe2[col_name].mode()[0])
    return dataframe2
```

```python
#In the categorical columns I'm' going to fill the missing values with the mode:

df = fill_nan_mode(df, "state")
df = fill_nan_mode(df, "response")
df = fill_nan_mode(df, "vehicle_class")
df = fill_nan_mode(df, "vehicle_size")
df = fill_nan_mode(df, "vehicle_type")

display(df.isna().sum()/len(df)*100)
```

```python
#Checking median and mean of the numeric columns:

display(df['months_since_last_claim'].median())
display(df['months_since_last_claim'].mean())

display(df['number_of_open_complaints'].median())
display(df['number_of_open_complaints'].mean())
```

```python
#Since these are supposed to be whole numbers i prefer to use the median to fill the missing values:

df = fill_nan_median(df, "months_since_last_claim")

df = fill_nan_median(df, "number_of_open_complaints")

display(df.isna().sum()/len(df)*100)
```

5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
df2=df.copy()

df2['month'] = pd.to_datetime(df['effective_to_date'], format='%m').dt.month_name()

grouped_table=df2.groupby(['state','month']).agg({"number_of_policies": ['count']}).reset_index()

grouped_table

pivot_table=pd.pivot(grouped_table, index='state', columns='month').reset_index()
display(pivot_table)

pivot_table.columns = pivot_table.columns.get_level_values(1)
pivot_table.columns.values[0] = 'state'
pivot_table.columns.values[1] = 'February'
pivot_table.columns.values[2] = 'January'
pivot_table
```

*Hint:*
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.* !SORT_VALUES (BY=['col1','col2'], ASCENDING=[FALSE, TRUE])!
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
df3=df.copy()

df3['month'] = pd.to_datetime(df['effective_to_date'], format='%m').dt.month_name()

# To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group.

df3=df3.groupby(['state','month']).agg({"number_of_policies": ['count']})

# Afterwards, you will need to sort the data by the count of policies sold in descending order.

df3=df3.sort_values(by=[('number_of_policies', 'count')], ascending=False).reset_index()

#Next, you will select the top 3 states with the highest number of policies sold.

df3=df3[(df3['state'] == 'California') | (df3['state'] == 'Oregon') | (df3['state'] == 'Arizona')]

df3
```
